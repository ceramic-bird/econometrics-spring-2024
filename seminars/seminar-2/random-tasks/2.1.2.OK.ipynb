```python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import seaborn as sns
import plotly.express as px
```

```python
# импорт данных из файла как DataFrame
df = pd.read_csv('sleep75.csv')
# размер датафрейма
df.shape
```

```python
# напечатаем датафрейм
df
```

```python
sns.set_theme(style='darkgrid')
sns.regplot(data=df, x='age', y='sleep', ci=None, scatter_kws={'s':5}, line_kws={'linestyle':'dashed', 'color':'r', 'linewidth':1})
```

```python
# спецификация модели через формулу
mod = smf.ols(formula='sleep~1+age', data=df)
# подгонка модели
res = mod.fit()
# параметры оптимальной прямой
res.params
```

```python
# спецификация модели через формулу
mod = smf.ols(formula='sleep~0+age', data=df)
# подгонка модели
res = mod.fit()
# параметры оптимальной прямой
res.params
```
